### 參考

## 範例
### 目錄中的全部ipynb

```cmd
jupyter nbconvert --to markdown notebook*.ipynb
``` 

如果要指定那些檔案,可以利用 configuration file(例如,mycfg.py),在裡面指定檔案:
mycfg.py內容
```
c = get_config()
c.NbConvertApp.notebooks = ["notebook1.ipynb", "notebook2.ipynb"]
```
然後利用命令
```
jupyter nbconvert --config mycfg.py
```
### 如果要指定輸出,下面是一個範例
```
jupyter nbconvert testconvert.ipynb --to markdown --NbConvertApp.output_files_dir="assets" --FilesWriter.build_directory="{notebook_name}.textbundle" --NbConvertApp.output_base="text"
```
執行結果: cmd 中執行上面的程式碼,會出現一個新的子目錄`{notebook_name}.textbundle`
```
C:.
│  testconvert.ipynb
│
└─{notebook_name}.textbundle
    │  text.md✔
    │
    └─assets
            text_0_0.png✔
```
⭕上面的程式碼修改成我要的,就是
```
jupyter nbconvert testconvert.ipynb --to markdown --NbConvertApp.output_files_dir="testconvert.files"  
```
執行結果
```
C:.
│  testconvert.ipynb
│  testconvert.md  ✔
│
└─testconvert.files
        testconvert_0_0.png ✔
```
