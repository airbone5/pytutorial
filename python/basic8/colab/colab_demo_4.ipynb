```python
from genericpath import isfile
from flask import Flask,render_template
from pyngrok import ngrok
from google.colab import userdata
import pyqrcode
from pyqrcode import QRCode
import os
from flask import request

auth_key=userdata.get('mykey')

app = Flask(__name__)

@app.route('/')
def home():
  return render_template('index.html')

@app.route('/handle_data', methods=['POST'])
def handle_data():
  projectpath = request.form['projectFilepath']
  #print(projectpath)
  #return projectpath

  files=os.listdir(f"static/{projectpath}")

  str=""
  for item in files:
    loc=os.path.join(f"static/{projectpath}",item)
    if os.path.isfile(loc):
      str=str+f"<br><a href=/static/{projectpath}/{item}>{item}</a>"
  return str



ngrok.set_auth_token(auth_key)
ngrok_tunnel = ngrok.connect(5000)
print('Public URL:', ngrok_tunnel.public_url)
url = pyqrcode.create(ngrok_tunnel.public_url)
url.png('myqr.png', scale=8)
app.run()
```
